```python
import math

def min_max_levels(n):
    min_levels = math.ceil(math.log2(n))
    max_levels = n
    return min_levels, max_levels

# Test cases
test_cases = [10, 35, 76, 345]

for n in test_cases:
    min_levels, max_levels = min_max_levels(n)
    print(f"Untuk n = {n} | Σlevel_minimum = {min_levels} | Σlevel_maksimum = {max_levels}")
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def generate_binary_trees(size):
    def generate_trees_helper(start, end):
        trees = []
        for i in range(start, end + 1):
            left_subtrees = generate_trees_helper(start, i - 1)
            right_subtrees = generate_trees_helper(i + 1, end)
            for left_tree in left_subtrees:
                for right_tree in right_subtrees:
                    root = TreeNode(i)
                    root.left = left_tree
                    root.right = right_tree
                    trees.append(root)
        return trees if trees else [None]

    return generate_trees_helper(1, size)

def print_tree(root, space=0):
    if root is None:
        return
    space += 5
    print_tree(root.right, space)
    print(" " * space + str(root.data))
    print_tree(root.left, space)

# Generate all possible binary trees with 5 nodes
all_trees = generate_binary_trees(5)

# Print all possible binary trees
print(f"Jumlah kemungkinan: {len(all_trees)}")
for idx, tree in enumerate(all_trees, start=1):
    print(f"Tree {idx}:")
    print_tree(tree)
    print()
```

```python
def max_nodes(h):
    return 2**h - 1

# Test cases
h_values = [3, 4, 5, 6]

for h in h_values:
    max_node_count = max_nodes(h)
    print(f"Untuk h = {h}, jumlah simpul maksimum adalah {max_node_count}.")
```
