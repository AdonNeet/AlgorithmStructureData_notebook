# Meng-ekstrak kata-kata dengan awalan ‘me’
Bukalah halaman web id.wikipedia. org/wiki/Indonesia. Salin seluruh tubuh artikel itu dan tuang ke Notepad. Simpan sebagai Indonesia.txt. Tugasmu adalah mengekstrak semua kata yang mempunyai awalan ‘me’, lalu menyimpannya dalam sebuah tuple. Perhatikan bahwa ‘me’ bisa muncul di awal,
tengah ataupun akhir kalimat.

```python
import re

with open('Indonesia.txt', 'r', encoding='utf-8') as f:
    teks = f.read()

p = r'\bme\w+'
ans = re.findall(p, teks)

print(ans)
```
