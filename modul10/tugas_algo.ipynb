```python
import time
import random
from timeit import timeit

def jumlahkan_cara_1(n):
    hasilnya = 0
    for i in range(1, n+1):
        hasilnya = hasilnya + i
    return hasilnya

h = jumlahkan_cara_1(10000)
for i in range(5): # mengulang lima kali
    print("Jumlah adalah %d, memerlukan %9.8f detik" % (h, timeit(lambda: jumlahkan_cara_1(10000), number=1)))


print(" ")
h = jumlahkan_cara_1(1000000)
for i in range(5): # mengulang lima kali
    print("Jumlah adalah %d, memerlukan %9.8f detik" % (h, timeit(lambda: jumlahkan_cara_1(1000000), number=1)))
```

```python
def jumlahkan_cara_2(n):
    return ( n*(n + 1) )/2

h = jumlahkan_cara_2(1000000)
for i in range(5): # mengulang lima kali
    print("Jumlah adalah %d, memerlukan %9.8f detik" % (h, timeit(lambda: jumlahkan_cara_2(1000000), number=1)))


print(" ")
h = jumlahkan_cara_2(2000000)
for i in range(5): # mengulang lima kali
    print("Jumlah adalah %d, memerlukan %9.8f detik" % (h, timeit(lambda: jumlahkan_cara_2(2000000), number=1)))
```

```python
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

print("Average Case")
for i in range(5):
    L = list(range(3000))
    random.shuffle(L) # Mengacak posisi elemen di list
    print("Mengurutkan %d bilangan, memerlukan %8.7f detik" % (3000, timeit(lambda: insertionSort(L), number=1)))

print(' ')
print("Worst Case")
for i in range(5):
    L = list(range(3000))
    L = L[::-1] # Membalik urutan elemen di list
    print("Mengurutkan %d bilangan, memerlukan %8.7f detik" % (3000, timeit(lambda: insertionSort(L), number=1)))

print(' ')
print("Best Case")
for i in range(5):
    L = list(range(3000))
    print("Mengurutkan %d bilangan, memerlukan %8.7f detik" % (3000, timeit(lambda: insertionSort(L), number=1)))
```

```python
import timeit
import matplotlib.pyplot as plt

## Ini fungsi yang akan diuji:
def test_append(n):
    lst = []
    for i in range(n):
        lst.append(i)

## Ini fungsi pengujinya:
def uji_append(n):
    ls = []
    jangkauan = range(1, n + 1)
    siap = "from __main__ import test_append"
    for i in jangkauan:
        # print('i =', i) # baris ini bisa dihidupkan atau dimatikan
        t = timeit.timeit("test_append(" + str(i) + ")", setup=siap, number=1)
        ls.append(t)
    return ls

## Pemanggilan pengujian:
n = 1000
LS = uji_append(n) # dari 1 sampai 1000.
## LS adalah list hasil uji kecepatan, dari n sedikit ke banyak.

## Menggambar grafik. Di bawah ini saja yang diulang saat menyesuaikan skala.
plt.plot(LS, label='Append Time') # Mem-plot hasil uji
skala = 1e-6 # Setel skala ini sesuai hasilmu.
plt.plot([x * skala for x in range(1, n + 1)], label='O(1)') # Grafik O(1) untuk pembanding.
plt.xlabel('Number of Appends')
plt.ylabel('Time (seconds)')
plt.title('Kompleksitas waktu pada list.append()')
plt.legend()
plt.show() # Tunjukkan plotnya
```

```python
import timeit
import matplotlib.pyplot as plt

## Ini fungsi yang akan diuji:
def insertAtBeginning(n):
    lst = []
    for i in range(n):
        lst.insert(0, i)

## Ini fungsi pengujinya:
def ujiInsertAtBeginning(n):
    ls = []
    jangkauan = range(1, n + 1)
    siap = "from __main__ import insertAtBeginning"
    for i in jangkauan:
        t = timeit.timeit("insertAtBeginning(" + str(i) + ")", setup=siap, number=1)
        ls.append(t)
    return ls

## Pemanggilan pengujian:
n = 1000
LS = ujiInsertAtBeginning(n) # dari 1 sampai 1000.
## LS adalah list hasil uji kecepatan, dari n sedikit ke banyak.

## Menggambar grafik. Di bawah ini saja yang diulang saat menyesuaikan skala.
plt.plot(LS, label='Insert di awal')
skala = max(LS) / (n * n) # Menyetel skala berdasarkan hasil pengujian.
plt.plot([x * x * skala for x in range(1, n + 1)], label='O(n^2)') # Grafik x^2 untuk pembanding.
plt.legend()
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Kompleksitas waktu pada insert() Method')
plt.show() # Tunjukkan plotnya
```

```python
import timeit
import matplotlib.pyplot as plt
import random

## Ini fungsi linear search yang akan diuji:
def linear_search(lst, target):
    for item in lst:
        if item == target:
            return True
    return False

## Ini fungsi pengujinya:
def ujiLinearSearch(n):
    ls = []
    for i in range(1, n + 1):
        L = list(range(i))
        target = -1 # Ensure the target is not in the list to get the worst case scenario
        t = timeit.timeit(lambda: linear_search(L, target), number=1)
        ls.append(t)
    return ls

## Pemanggilan pengujian:
n = 1000
LS = ujiLinearSearch(n) # dari 1 sampai 1000.
## LS adalah list hasil uji kecepatan, dari n sedikit ke banyak.

## Menggambar grafik. Di bawah ini saja yang diulang saat menyesuaikan skala.
plt.plot(LS, label="Linear Search Timing") # Mem-plot hasil uji
skala = max(LS) / (n * n) # Adjust the scale to fit the comparison curve
plt.plot([x * x * skala for x in range(1, n + 1)], label="O(n^2) Comparison") # Grafik x^2 untuk pembanding.
plt.legend()
plt.show() # Tunjukkan plotnya
```
